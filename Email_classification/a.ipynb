```python

balance_df.reset_index(drop=True,inplace=True)
```

```python

import re  # regular expression  
```

```python

balance_df['email_messages'][0]
```

```python

for i in range(len(balance_df)): 
    lower_message = balance_df['email_messages'][i].lower()
    clean_message = re.sub('[^a-zA-Z ]',"",lower_message)  # substitud  
    balance_df['email_messages'][i] = clean_message 
    
```

```python

balance_df['email_messages'][0]
```

```python
x  =  balance_df['email_messages'] 
y  =  balance_df['labels']
```

```python

X = x.values  # changed into numpy array 
```

```python

from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer  
## to convert text data to numerical representation 
```

```python
bow_obj =  CountVectorizer(max_features=1500,stop_words='english')
```

```python
# from nltk.corpus import stopwords 
# len(stopwords.words('english')) 
```

```python

x_transformed = bow_obj.fit_transform(X)
x_transformed
```

```python

X.shape 
```

```python

X2 = x_transformed.toarray() 
```

```python
dt={'spam':1,'ham':0}
y=y.map(dt)
```

```python
import random
```

```python
ls=[25,50,75,100]
random.shuffle(ls)
```

```python
ls
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X2,y, test_size=0.10)

# print("x_train shape:", x_train.shape)
# print("x_test shape:", x_test.shape)
# print("y_train shape:", y_train.shape)
# print("y_test shape:", y_test.shape)
```

```python
x_train[0][0:500]
```

```python
# algorithmn training:
# This problem is a binary classification task. The goal is to predict whether an email message is spam (1) or not spam (0). The dataset consists of a collection of SMS messages labeled as either "spam" or "ham". 
```

naive bayes

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
# 3 classes,
# bayes theorm deals with conditional probability
# Naive bayes algorithmn:It is mainly used in text classification,it will work with high diemnsional training dataset
# It is based to Bayes theorem used for classification problems
# some eg of Naive Bayes are:
# spam filtration,sentimental analysis,classifying articles
```

```python
# object
mnb=MultinomialNB()
bnb=BernoulliNB()

```

```python
mnb.fit(x_train,y_train)
bnb.fit(x_train,y_train)
print("successfully trained!")
```

```python
# score checkout
print("MultinomialNB")
print("Training score:",mnb.score(x_train,y_train)*100)
print("Testing score:",mnb.score(x_test,y_test)*100)

print("BernoulliNB")
print("Training score:",bnb.score(x_train,y_train)*100)
print("Testing score:",bnb.score(x_test,y_test)*100)
```

```python
# prediction
prediction=bnb.predict(x_test)
prediction
```

```python
# actual
np.array(y_test)
```

```python
evaluate_df=pd.DataFrame({'Actual':np.array(y_test),'prediction':prediction})
evaluate_df.head(40)
```

```python
from sklearn.metrics import confusion_matrix,classification_report


matrix = confusion_matrix(evaluate_df['Actual'],evaluate_df['prediction'])
matrix
# plot
# 90 ARE CORRECTED TRUE PoSTIVE
# 69 ARE TRUE NEGATIVE
# O False POSITIVE
# 10 false negative
```

```python
sns.heatmap(matrix,annot=True,fmt='d')
plt.title('confusion_matrix')
plt.axis('off')
plt.show()
```

```python
print(classification_report(evaluate_df['Actual'],evaluate_df['prediction']))
```
