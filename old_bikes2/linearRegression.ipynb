```python
dt = {'First Owner':1,'Second Owner':2,
      'Third Owner':3,"Fourth Owner Or More":4}

dt
```

```python
df['owner'] = df['owner'].map(dt)  # single time execute 
```

```python
df
```

```python
df['brand'].value_counts()
```

```python
bike_brand =  list(df['brand'].value_counts().head(16).index)
bike_brand
```

```python
# (df['brand'] == "Bajaj") | (df['brand'] == "Royal Enfield") 
df2 =  df[df['brand'].isin(bike_brand)]
df2.head()
```

```python
df2['brand'].value_counts() 
```

```python
ls = list(df2['brand'].value_counts().keys())
```

```python
for i in range(len(df2['brand'].value_counts())):
    dt[ls[i]]=i+1
```

```python
df2['brand'] =  df2['brand'].map(dt)
```

```python
df2.drop(['bike_name','city'],axis='columns',inplace=True)
```

```python
df2.head()
```

```python
x = df2.drop('price',axis='columns')   # independents variables 
y = df2[['price']]  # dependent variable or target varible 
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# importing linearregression algorithm class 
from sklearn.linear_model import LinearRegression 
```

```python
#instance creation 
lnr = LinearRegression()
```

```python
## algorithm train 
lnr.fit(x_train,y_train)
```

```python

## Model evaluation 
print("Training score  ",lnr.score(x_train,y_train)*100) 
print("Testing score  ",lnr.score(x_test,y_test)*100 )
```

```python
x_test.shape[0]
# total 1462 bikes 
```

```python
### Prediction or testing phase 

prediction = lnr.predict(x_test)
```

```python
prediction.shape
```

```python
comparedf =pd.concat([x_test,y_test],axis=1)
```

```python
comparedf['prediction'] = prediction 
```

```python
comparedf.head()
```

```python
comparedf['residual'] =  comparedf['price']  - comparedf['prediction']  
```
